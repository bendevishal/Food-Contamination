```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, measure, morphology
from skimage.color import rgb2gray
from skimage.segmentation import clear_border
from scipy import ndimage as ndi

# 1. Load and Display the Image
def load_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.title('Original Image')
    plt.axis('off')
    plt.show()
    return image_rgb

# 2. Preprocess the Image
def preprocess_image(image_rgb):
    gray_image = rgb2gray(image_rgb)
    blurred_image = filters.gaussian(gray_image, sigma=1.0)
    plt.imshow(blurred_image, cmap='gray')
    plt.title('Grayscale and Blurred Image')
    plt.axis('off')
    plt.show()
    return blurred_image

# 3. Edge Detection with Canny
def detect_edges(blurred_image):
    # Use Canny edge detection instead of Sobel to avoid detecting darkness as edges
    low_threshold = 0.1  # lower threshold for edge detection
    high_threshold = 0.3  # upper threshold for edge detection
    edges = cv2.Canny((blurred_image * 255).astype(np.uint8), int(low_threshold * 255), int(high_threshold * 255))

    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edge Detection')
    plt.axis('off')
    plt.show()
    return edges

# 4. Adaptive Thresholding and Morphological Operations
def apply_threshold(edges):
    # Adaptive thresholding to account for uneven lighting/darkness
    binary_image = cv2.adaptiveThreshold(edges, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                         cv2.THRESH_BINARY, 11, 2)

    # Morphological operations to clean the binary image
    binary_image = morphology.binary_opening(binary_image, morphology.disk(2))
    binary_image = morphology.binary_closing(binary_image, morphology.disk(2))
    binary_image = clear_border(binary_image)  # Remove artifacts connected to the image border

    plt.imshow(binary_image, cmap='gray')
    plt.title('Binary Image after Morphological Operations')
    plt.axis('off')
    plt.show()
    return binary_image

# 5. Label and Measure Regions
def label_and_measure(binary_image):
    labeled_image, num_labels = ndi.label(binary_image)
    props = measure.regionprops(labeled_image)

    plt.imshow(labeled_image, cmap='nipy_spectral')
    plt.title(f'Labeled Regions ({num_labels} regions)')
    plt.axis('off')
    plt.show()

    return labeled_image, props

# 6. Contaminant Detection and Highlighting
def highlight_contaminants(image_rgb, labeled_image, props):
    output_image = image_rgb.copy()

    for prop in props:
        if prop.area > 50:  # Example criterion for contamination
            minr, minc, maxr, maxc = prop.bbox
            cv2.rectangle(output_image, (minc, minr), (maxc, maxr), (255, 0, 0), 2)

    plt.imshow(output_image)
    plt.title('Contaminants Highlighted')
    plt.axis('off')
    plt.show()

# 7. Main function to run the project
def main(image_path):
    image_rgb = load_image(image_path)
    blurred_image = preprocess_image(image_rgb)
    edges = detect_edges(blurred_image)
    binary_image = apply_threshold(edges)
    labeled_image, props = label_and_measure(binary_image)
    highlight_contaminants(image_rgb, labeled_image, props)

# Run the project with your image path
if __name__ == "__main__":
    image_path = '/content/pic2.jpg'  # Replace with the path to your image
    main(image_path)
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology
from skimage.color import rgb2gray

# 1. Load and Display the Image
def load_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.title('Original Image')
    plt.axis('off')
    plt.show()
    return image_rgb

# 2. Preprocess the Image (Convert to grayscale and blur)
def preprocess_image(image_rgb):
    gray_image = rgb2gray(image_rgb) * 255  # Convert to grayscale and scale to [0, 255]

    # Apply Gaussian blur to smooth the image
    blurred_image = cv2.GaussianBlur(gray_image.astype(np.uint8), (5, 5), 0)

    plt.imshow(blurred_image, cmap='gray')
    plt.title('Grayscale and Blurred Image')
    plt.axis('off')
    plt.show()

    return blurred_image

# 3. Detect Black Spots (Contaminants) Using Adaptive Thresholding
def detect_black_spots(blurred_image):
    # Threshold for detecting black spots (low intensity regions)
    _, black_spots = cv2.threshold(blurred_image, 50, 255, cv2.THRESH_BINARY_INV)

    # Morphological operations to remove small noise and connect small objects
    black_spots_cleaned = morphology.binary_opening(black_spots, morphology.disk(3))
    black_spots_cleaned = morphology.binary_closing(black_spots_cleaned, morphology.disk(3))

    plt.imshow(black_spots_cleaned, cmap='gray')
    plt.title('Detected Black Spots (Contaminants)')
    plt.axis('off')
    plt.show()

    return black_spots_cleaned

# 4. Detect White Spots (Contaminants) Using Otsu Thresholding
def detect_white_spots(blurred_image):
    # Otsu's method to threshold white spots
    _, white_spots = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Morphological operations to clean up noise and isolate white spots
    white_spots_cleaned = morphology.binary_opening(white_spots, morphology.disk(3))
    white_spots_cleaned = morphology.binary_closing(white_spots_cleaned, morphology.disk(3))

    plt.imshow(white_spots_cleaned, cmap='gray')
    plt.title('Detected White Spots (Contaminants)')
    plt.axis('off')
    plt.show()

    return white_spots_cleaned

# 5. Draw Contours for Detected Spots (Black and White Contaminants)
def highlight_contaminants(image_rgb, black_spots, white_spots):
    output_image = image_rgb.copy()

    # Find contours for black spots (contaminants)
    contours_black, _ = cv2.findContours(black_spots.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours for black spots (in red)
    for contour in contours_black:
        if cv2.contourArea(contour) > 50:  # Ignore small noise
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(output_image, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Red bounding box
            cv2.drawContours(output_image, [contour], -1, (255, 0, 0), 2)  # Red contour

    # Find contours for white spots (contaminants)
    contours_white, _ = cv2.findContours(white_spots.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours for white spots (in green)
    for contour in contours_white:
        if cv2.contourArea(contour) > 50:  # Ignore small noise
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(output_image, (x, y), (x + w, y + h), (0, 255, 0), 2)  # Green bounding box
            cv2.drawContours(output_image, [contour], -1, (0, 255, 0), 2)  # Green contour

    plt.imshow(output_image)
    plt.title('Detected Contaminants Highlighted (Black and White Spots)')
    plt.axis('off')
    plt.show()

# 6. Main function to run the project
def main(image_path):
    image_rgb = load_image(image_path)
    blurred_image = preprocess_image(image_rgb)
    black_spots = detect_black_spots(blurred_image)
    white_spots = detect_white_spots(blurred_image)
    highlight_contaminants(image_rgb, black_spots, white_spots)

# Run the project with your image path
if __name__ == "__main__":
    image_path = '/content/image2.jpg'  # Replace with the path to your image
    main(image_path)
```
